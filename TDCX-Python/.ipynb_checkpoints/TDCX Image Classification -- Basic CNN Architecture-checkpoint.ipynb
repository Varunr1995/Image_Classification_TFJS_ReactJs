## Importing Libraries

```python
import numpy as np
import os
from os import path
import PIL
import matplotlib.pyplot as plt
import cv2
from sklearn import preprocessing
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from tensorflow.keras.models import load_model
import tensorflowjs as tfjs
```

```python
train = "/Users/varun/Desktop/TDCX/train/train/"
test = "/Users/varun/Desktop/TDCX/test/test/"
```

```python
# Splitting the data as per the file format in to lables and jpg

train_labels = []
test_labels = []

train_path = []
test_path = []
```

```python
# Loop running to split the data asper the file format.

for filename in os.listdir(train):
    if(filename.split('.')[1]=="jpg"):
        train_labels.append(filename.split('_')[0])
        train_path.append(os.path.join(train, filename))
        
for filename in os.listdir(test):
    if(filename.split('.')[1]=='jpg'):
        test_labels.append(filename.split('_')[0])
        test_path.append(os.path.join(test, filename))
```

```python
train_label_unique = np.unique(np.array(train_labels))
test_label_unique = np.unique(np.array(test_labels))

print("Unique labels in train dataset are: ", train_label_unique)
print("Number of jpg images in train dataset are: ", len(train_path))

print("\nUnique labels in test dataset are: ", test_label_unique)
print("Number of jpg images in train dataset are: ", len(test_path))
```

## Data Verification

```python
image = cv2.imread(train_path[0])

rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

plt.imshow(rgb_img)
plt.title("Label: " + train_labels[0])
plt.axis('off')
plt.show()
```

```python
image = cv2.imread(test_path[0])

rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

plt.imshow(rgb_img)
plt.title("Label: " + test_labels[0])
plt.axis('off')
plt.show()
```

## Splitting Data in to Train & Test DataSets

```python
x_train = []

for path in train_path:
    img = cv2.imread(path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    final_img =  cv2.resize(rgb_img, (50,50))
    x_train.append(final_img)

x_train = np.array(x_train)
```

```python
x_test = []

for path in test_path:  
    img = cv2.imread(path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    final_img =  cv2.resize(rgb_img, (50,50))
    x_test.append(final_img)

x_test = np.array(x_test)
```

```python
fruit_names = {}

for i in range(len(train_label_unique)):
    fruit_names[train_label_unique[i]] = i
    
fruit_names
```

```python
# storing the values in a temporary list
train_temp = []
test_temp = []

# all the fruits names are being mapped
for label in train_labels:
    train_temp.append(fruit_names.get(label))

for label in test_labels:
    test_temp.append(fruit_names.get(label))

print("Length of train data: ", len(train_temp))
print("Length of test data: ", len(test_temp))
```

```python
y_train = keras.utils.to_categorical(train_temp, 4)
y_test = keras.utils.to_categorical(test_temp, 4)
```

```python
print("Length of x_train: ", len(x_train))
print("Length of x_test: ", len(x_test))
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python
print("x_train shape: ", x_train.shape)
print("x_test shape: ", x_test.shape)

print("\ny_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

## Developing Basic Sequential model

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), padding='same', input_shape=(50, 50, 3), activation="relu"))
model.add(Conv2D(32, (3, 3), activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same', activation="relu"))
model.add(Conv2D(64, (3, 3), activation="relu"))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(4, activation="softmax"))
```

```python
# Model Compilation
model.compile(
    loss='categorical_crossentropy',
    optimizer="adam",
    metrics=['accuracy']
)

model.summary()
```

```python
# Model Training
history = model.fit(
    x_train,
    y_train,
    batch_size=16,
    epochs=150,
    validation_split=0.2,
    shuffle=True,
    verbose=1
)
```

## Result Analysis

```python
def plot_accuracy_loss(history):
    plt.figure(figsize = (30, 20))
    plt.subplot(221)
    plt.plot(history.history['accuracy'], 'bo--', label = 'Accuracy')
    plt.plot(history.history['val_accuracy'], 'ro--', label = 'validation_Accuracy')
    plt.title("Train_Accuracy Vs Validation_Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    
    plt.subplot(222)
    plt.plot(history.history['loss'], 'bo--', label = 'Loss')
    plt.plot(history.history['val_loss'], 'ro--', label = 'Validation_Loss')
    plt.title("Train_Loss Vs Validation_Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.figure(figsize = (30, 20))
    
    plt.show()
```

```python
plot_accuracy_loss(history)
```

```python
tfjs.converters.save_keras_model(model, 'tf_js')
# model.save("TDCX.h5")
```

## Model Evaluation

```python
score, accuracy = model.evaluate(x_test, y_test)
print('Test score achieved:', score)
print('Test accuracy achieved:', accuracy)
```
